```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el DataFrame
in_g = df
# Calcular la matriz de correlación
corr_matrix = in_g.corr()

# Filtrar la correlación con la columna específica
target_col = '% Población de 15 años o más analfabeta '
correlations = corr_matrix[target_col].drop(target_col)

# Crear mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlations.to_frame(), annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlación con {}'.format(target_col))
plt.show()
```

#h) Desarrolla un nuevo DataFrame con indicadores interesantes por estado que se obtengan de los datos a nivel municipal y que pudieran ser de importancia a la hora de definir políticas públicas. Se va a calificar la originalidad de los indicadores. Justifica las decisiones que tomes, y guarda el nuevo dataframe en formato parquet.

Aqui presento y justifico los indicadores que eligiré en el dataframe que considero relevantes a la hora de definir politicas publicas:

1. Porcentaje de población de 15 años o más analfabeta por estado.<br>
Este indicador proporciona información sobre el nivel de analfabetismo en cada entidad federativa. El analfabetismo está estrechamente relacionado con la falta de acceso a la educación y puede ser un indicativo de la calidad y la disponibilidad de los servicios educativos. 

2. Porcentaje de ocupantes en viviendas particulares sin servicios básicos por estado.<br>
Este indicador refleja la calidad de vida y el acceso a servicios básicos en cada estado. Las viviendas sin drenaje, sin energía eléctrica o sin agua entubada indican una falta de infraestructura básica que puede afectar la salud, la seguridad y el bienestar de la población.

3. Porcentaje de viviendas particulares con piso de tierra por estado.<br>
Este indicador está relacionado con las condiciones de vivienda y puede ser un reflejo de la pobreza y la falta de desarrollo. Las viviendas con piso de tierra están asociadas con mayores riesgos de enfermedades, higiene deficiente y menor calidad de vida en general.

4. Porcentaje de población en localidades con menos de 5,000 habitantes por estado.<br>
Este indicador se enfoca en las áreas rurales y la distribución de la población. Las localidades rurales suelen tener características y necesidades diferentes a las áreas urbanas. 

5. Porcentaje de población ocupada con ingresos menores a 2 salarios mínimos por estado.<br>
Este indicador destaca la situación económica de la población trabajadora. Un porcentaje alto de población ocupada con ingresos bajos puede indicar la existencia de desigualdades salariales, falta de empleos de calidad y dificultades para cubrir las necesidades básicas.


```python
# Porcentaje de población de 15 años o más analfabeta por estado: 
import pandas as pd

# Carga el DataFrame con los datos
ins_g=df

# Agrupa por estado y calcula el promedio del porcentaje de población analfabeta
ins_g_estado = df.groupby('Nombre de la entidad').agg({'Población total': 'sum',
                                                        '% Población de 15 años o más analfabeta ': 'sum',
                                                        '% Ocupantes en viviendas particulares sin drenaje ni excusado': 'sum',
                                                        '% Ocupantes en viviendas particulares sin energía eléctrica': 'sum',
                                                        '% Ocupantes en viviendas particulares sin agua entubada': 'sum',
                                                        '% Ocupantes en viviendas particulares con piso de tierra': 'sum',
                                                        '% Viviendas particulares con hacinamiento': 'sum',
                                                        '% Población en localidades con menos de 5 000 habitantes': 'sum',
                                                        '% Población ocupada con ingresos menores a 2 salarios mínimos':'sum'})

# Elimina la columna 'municipio' del nuevo DataFrame
ins_g_estado = ins_g_estado.reset_index()

# Imprime el nuevo DataFrame 
ins_g_estado
```
